```python
import matplotlib.pyplot as plt
import seaborn as sns
from math import log2

def entropy(t):
    return -sum([p*log2(p) for p in t if p > 0])

N = 10
a = list(range(0, N+1)) # 0 to 10
b = list(range(N, -1, -1)) # 10 to 0
dist = list(zip(a, b))


def graph(dist):
    fig, ax = plt.subplots()
    ax.set_title("Entropy of a 2-class distribution")
    ax.set_xlabel("Probability")
    ax.set_ylabel("Entropy")

    pairs = [(a/(a+b), entropy((a/(a+b), b/(a+b)))) for a, b in sorted(dist)]
    # Separate the probabilities and entropies for plotting
    probabilities, entropies = zip(*pairs)

    # Plot the line
    ax.plot(probabilities, entropies, 'b-')

    for a, b in dist:
        t = [a/(a+b), b/(a+b)]
        ax.plot(t[0], entropy(t), 'ro')
        ha = 'right' if t[0] < .5 else 'left'
        if t[0] == .5:
            ha = 'center'
        va = 'bottom' if entropy(t) < .5 else 'top'
        ax.text(t[0], entropy(t)-.03, f"({a}, {b})", ha=ha, va=va )
    plt.show()
    # save as an svg
    # fig.savefig("entropy.svg")
graph(dist)
```
