```python
import pandas as pd
from nycschools import schools, geo, ui


```

```python
df = schools.load_school_demographics()
df.ay.unique()
cols = [c for c in df.columns if c.startswith(
    'grade') or c.endswith('_n') or c.endswith('_pct') or c in ["total_enrollment", "eni"]]
cols
```

```python
data = df.sort_values("ay")
diff_cols = [c+"_diff" for c in cols]
data[diff_cols] = df[cols].diff()
data[["dbn","school_name", "ay","white_pct_diff"]].sort_values("white_pct_diff", ascending=False).head(10)
```
