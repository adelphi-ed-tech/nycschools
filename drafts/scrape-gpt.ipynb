```python
import os
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from IPython.display import Markdown as md
```

```python
link_path = "/home/mxc/Videos/AAL-ChatGPT/trial-links.csv"
links = pd.read_csv(link_path)

links
```

```python
options = Options()
options.add_argument("--headless")
driver = webdriver.Chrome(options=options)

def get_trial(row):
    try:
        driver.get(row.Link)

        WebDriverWait(driver, 10).until( EC.presence_of_element_located( (By.CSS_SELECTOR, "article")) )

        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')

        articles = soup.find_all('article')

        msgs = []
        for i, article in enumerate(articles):
            trial = {
                "trial_id": row.Trial_ID,
                "condition": row.Condition,
                "passage": row.Passage,
                "msg_num": i+1
            }
            txt = article.get_text()
            if txt.startswith("You said:"):
                trial.update({"speaker": "user", "msg": txt[len("You said:"):]})
            else:
                trial.update({"speaker": "gpt", "msg": txt[len("ChatGPT said:"):]})
            msgs.append(trial)
        
        print(row["Trial_ID"])
        return msgs
    except Exception as e:
        print(e)
        print(row.Condition)
        print(row.Link)
        return []
t = list(links.apply(get_trial, axis=1))
# t = list(links[links.Trial_ID == 17].apply(get_trial, axis=1))
driver.quit()
df = pd.concat([pd.DataFrame(trial) for trial in t])

df
```

```python
driver = webdriver.Chrome(options=options)
x = list(links[links.Trial_ID == 17].apply(get_trial, axis=1))
driver.quit()
x
```

```python
# links[links.Trial_ID == 17].apply(get_trial, axis=1)
y = pd.concat([pd.DataFrame(trial) for trial in x])
combo = pd.concat([df, y])
combo = combo.sort_values(by=["trial_id", "msg_num"])
combo.loc[combo.condition.str.startswith("Chi"), "l1"] = "ch"
combo.loc[combo.condition.str.startswith("Spa"), "l1"] = "es"
combo.tail(10)
```
