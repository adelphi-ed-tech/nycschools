Querying & Selecting
====================

In this Notebook we go over some of the ways of querying data in a dataframe.

- [Subsetting Data](https://datacarpentry.org/python-ecology-lesson/03-index-slice-subset/#subsetting-data-using-criteria) using criteria
- [`query()`](https://www.w3schools.com/python/pandas/ref_df_query.asp) method of [`DataFrame`](https://www.w3schools.com/python/pandas/pandas_dataframes.asp) to search data
- using [`isin()`](https://www.w3schools.com/python/pandas/ref_df_isin.asp), [`isnull()`](https://www.w3schools.com/python/pandas/ref_df_isnull.asp)
- [`drop_duplicates()`](https://www.w3schools.com/python/pandas/pandas_cleaning_duplicates.asp)


```python
# import schools from the nycschool package
from nycschools import schools, exams

# load the demographic data into a `DataFrame` called df
df = schools.load_school_demographics()
```

```python
# we already saw how to get a subset of columns
# let's get just these columns from our data
cols = ["dbn", "ay", "school_name", "district", "poverty_pct", "ell_pct", "swd_pct"]

#by using the variable declaration below of 'df = df[cols]', we are now "updating" our 'df' to only have the columns specified in 'cols'
df = df[cols]
df
```

Selecting just one year
---------------------------------
We can see that our data contains multiple years of data for each school.
We can filter or query the data to just get a single year.

To do this, we will use the slice notation similary to above, however,
instead of a list of columns, we have a Boolean expression using
one of the columns in our data set.

`df[df.ay == 2020]` returns only the rows where `ay` equals `2020`


```python
ay_2020 = df[df.ay == 2020]
ay_2020.head()
```

Compound expressions
----------------------------------
We can use Boolean operators to make more complex queries. 
`pandas` uses the `&` operator for Boolean **and**
and the `|` operator for Boolean **or**. Note that you should wrap
your equality expressions inside of parentheses.
<br><br>
Notes on all Python Operators can be found [here](https://www.w3schools.com/python/python_operators.asp).
<br><br>

To write bug-free code, we do not recommend mixing `and` and `or`
clauses in the same query.

Below are two examples:

- `df[(df.district == 13) & (df.ay == 2020)]`
   find data for schools in district 13 _and_ academic year 2020-21
- `df[(df.poverty_pct > .9) | (df.ell_pct >= .4)]`
   find school data where the school's poverty percent is greater than 90% _or_
   the percent of students classified as English Language Learners is greater
   than or equal to 40%


```python
df[(df.district == 13) & (df.ay == 2020)].head()
```

```python
df[(df.poverty_pct > .9) | (df.ell_pct >= .4)].head()
```

```python
# this works, but it is confusing and error prone
df[(df.ay == 2020)  & ((df.poverty_pct > .9) | (df.ell_pct >= .4))].head()
```

```python
# rather than mixing | and &, we can write our code more clearly in two lines
data = df[df.ay == 2020]
data = data[(data.poverty_pct > .9) | (data.ell_pct >= .4)]
data.head()
```

Finding a subset of data
-----------------------------------
`DataFrame` has a useful function called `isin()` that lets us filter data on a `Series`
if the value is in a set of values. <br>First, we'll do a simple example where we find
all of the schools in a district `[3,8,15]`

```python
districts = [3, 8, 15]
data = df[df.district.isin(districts)]
data[["dbn", "district", "school_name"]]
```

Using `drop_duplicates`
-----------------------------------------------------
In the display of the table above, since we chose not to show all of the columns, we have repeated columns. Sometimes we will want just the unique results. We saw that `unique()` works for a `Series`. It doesn't work for a dataframe, however.

Luckily, we have `drop_duplicates()` which serves a similar function...see:


```python
data[["dbn", "district", "school_name"]].drop_duplicates()
```

Filtering missing data
-------------------------------
You will not always have complete data, and sometimes you need to find rows with missing records. You may want to update them with some default or derived values, or, you may simply want to exclude them from your data.

In this example, we will use `nycschools` to load the grade 3-8 ELA NYS exams. We will then filter out all of the rows (using the [`isnull()`](https://www.w3schools.com/python/pandas/ref_df_isnull.asp) method) that don't have a score. Some rows are null because not enough students took the exam. Others may be missing data because their students were exempt from the exam or other reasons.

```python
from nycschools import exams
ela = exams.load_ela()
print(f"There are {len(ela):,} total records in ela.")
ela.head()
```

```python
# Here we declare a variable 'missing', 
# by looking into the ela dataframe created above and returning all rows for which the 'mean_scale_score' is a null value [using the isnull() method]

missing = ela[ela.mean_scale_score.isnull()]
print(f"There are {len(missing):,} records in ela with null values for mean_scale_score. NaN means 'Not a Number'")

missing.head()
```

Conversely, we can use the [`notnull()`](https://www.w3schools.com/python/pandas/ref_df_notnull.asp) method in a similar manner to return only those rows for which 'mean_scale_score' is not null (NaN).

```python
# Here we declare a variable 'found', 
# by looking into the ela dataframe created above and returning all rows for which the 'mean_scale_score' is a defined value [using the notnull() method]

found = ela[ela.mean_scale_score.notnull()]
print(f"There are {len(found):,} records in ela with defined values for mean_scale_score.")

found.head()
```
